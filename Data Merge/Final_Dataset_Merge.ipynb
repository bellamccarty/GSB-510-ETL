```python
import pandas as pd

# Load datasets
airbnb_df = pd.read_csv(
    r"C:\Users\isabe\PycharmProjects\GSB-520-ETL\Clean Data\cleaned_airbnb_data.csv")
tourism_df = pd.read_csv(
    r"C:\Users\isabe\PycharmProjects\GSB-520-ETL\Clean Data\tourism_cleaned.csv")
zillow_df = pd.read_csv(
    r"C:\Users\isabe\PycharmProjects\GSB-520-ETL\Clean Data\zillow_cleaned.csv")

# Extract Year from dates
airbnb_df["Year"] = pd.to_datetime(airbnb_df["Date Last Review"]).dt.year
zillow_df["Year"] = pd.to_datetime(zillow_df["Date"]).dt.year

# Aggregate Airbnb data by City + Year
airbnb_agg = airbnb_df.groupby(["City", "Year"]).agg({
    "Room Price": "mean",
    "Number of reviews": "sum",
    "Availibility": "mean"
}).reset_index()

# Rename columns for clarity
airbnb_agg.rename(columns={
    "Room Price": "Avg_Room_Price",
    "Number of reviews": "Total_Reviews",
    "Availibility": "Avg_Availability"
}, inplace=True)

# Aggregate Zillow data by City + Year
zillow_agg = zillow_df.groupby(["City", "State", "Year"]).agg({
    "New_Renter_Affordability": ["mean", "min", "max"]
}).reset_index()
zillow_agg.columns = ["City", "State", "Year", "Avg_Affordability", "Min_Affordability", "Max_Affordability"]

# Merge Airbnb & Tourism (Outer Join)
airbnb_tourism_merged = pd.merge(airbnb_agg, tourism_df, on="City", how="outer")

# Merge with Zillow (Outer Join on City, State, Year)
final_merged_df = pd.merge(airbnb_tourism_merged, zillow_agg, on=["City", "State", "Year"], how="outer")
```
